```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define the frequency response for MA filter

```python

def H_MA(M1, M2, omega):
    N = M1 + M2 + 1
    H = (1 / N) * (np.sin((omega * N) / 2) / np.sin(omega / 2)) * np.exp(-1j * omega * (M1 - M2) / 2)
    return H

# Define the parameters
M1_values = [0, 0]
M2_values = [1, 4]
```

```python
# Define the frequency range
omega = np.linspace(-np.pi, np.pi, 1000)
```

## Plotting the magnitude and phase responses

```python

plt.figure(figsize=(14, 12))

for i, (M1, M2) in enumerate(zip(M1_values, M2_values)):
    H = H_MA(M1, M2, omega)
    
    # Magnitude response
    plt.subplot(2, 2, i*2 + 1)
    plt.plot(omega, np.abs(H))
    plt.title(f'Magnitude Response for M1 = {M1}, M2 = {M2}')
    plt.xlabel('Frequency (rad/sample)')
    plt.ylabel('Magnitude')
    plt.grid(True)
    
    # Phase response
    plt.subplot(2, 2, i*2 + 2)
    plt.plot(omega, np.angle(H))
    plt.title(f'Phase Response for M1 = {M1}, M2 = {M2}')
    plt.xlabel('Frequency (rad/sample)')
    plt.ylabel('Phase (radians)')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
### Importing necessary Modules
import numpy as np
import matplotlib.pyplot as plt

### Defining the frequency range
omega = np.linspace(-np.pi, np.pi, 1000)

### Define the values of alpha
alpha_values = [9/10, 1/2, 1/4]
### Computing the frequency response for each alpha
H = lambda alpha, omega: 1 / (1 - alpha * np.exp(-1j * omega))
## Plotting the magnitude response

plt.figure(figsize=(12, 8))

for alpha in alpha_values:
    H_alpha = H(alpha, omega)
    plt.plot(omega, np.abs(H_alpha), label=f'α = {alpha}')

plt.title('Magnitude Response of the System for Different α Values')
plt.xlabel('Frequency (rad/sample)')
plt.ylabel('Magnitude')
plt.legend()
plt.grid(True)
plt.show()

## Observations
""" For 
𝛼 = 9/10:
The magnitude response is highest, indicating a strong frequency response.
The system has a significant peak around \n 
ω=0, suggesting that it acts as a low-pass filter with a high gain near the DC component.
For \n
𝛼 = 1/2 :
The magnitude response is moderate.
The peak is lower compared to alpha = 9/10 indicating a less pronounced low-pass filter effect.
For \n
𝛼 = 1/4 :
The magnitude response is the lowest among the three values of 𝛼 . 
The system's response is more spread out, with a smaller peak near 
𝜔=0 indicating a more balanced frequency response with lower overall gain."""
```
